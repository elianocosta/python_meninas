```python
# Esses dois comandos são usados em Python para importar bibliotecas de visualização de dados. Vamos por partes:
# import seaborn as sns
#Significado: Está importando a biblioteca Seaborn e atribuindo o apelido (alias) sns para facilitar o uso nos comandos seguintes.
# Para que serve: O Seaborn é uma biblioteca baseada no Matplotlib, que facilita a criação de gráficos estatísticos mais bonitos e informativos, como:
# gráficos de barras /gráficos de dispersão (scatterplots)/ mapas de calor (heatmaps)/ boxplots
#import matplotlib.pyplot as plt
# Significado: Está importando o módulo pyplot da biblioteca Matplotlib com o apelido plt.
# Para que serve: O pyplot fornece funções para criar e personalizar gráficos.
# Ele é como uma "base" de visualização, e o Seaborn usa o Matplotlib por trás dos panos.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Esse comando está carregando uma planilha do Excel para dentro de um DataFrame do Pandas, uma estrutura de dados muito usada para análise em Python.
#  Parte por parte:
# df: É o nome da variável onde os dados serão armazenados. df é abreviação de DataFrame.
# pd.read_excel(...):
# pd é o apelido da biblioteca Pandas, que precisa ter sido importada antes assim:
#import pandas as pd
# read_excel é uma função do Pandas que lê um arquivo .xlsx (Excel) e transforma os dados em uma tabela manipulável em Python (um DataFrame).
# 'resumo_semanal_lpc_2025-03-30_2025-04-05.xlsx':
# Esse é o nome do arquivo Excel que você está abrindo. Ele precisa estar na mesma pasta do código, a menos que você coloque o caminho completo.
# header=9:
# Isso significa que a linha 10 da planilha (lembre que o Python começa a contar do 0) será usada como título das colunas.
# Útil quando a planilha tem títulos, datas ou explicações nas primeiras linhas, e os dados reais começam mais abaixo.
# Em resumo:
# Esse comando está dizendo:
# “Leia a planilha chamada resumo_semanal_lpc_2025-03-30_2025-04-05.xlsx, use a linha 10 como título das colunas, e guarde os dados na variável df.”
df = pd.read_excel('resumo_semanal_lpc_2025-03-30_2025-04-05.xlsx', header=9)
```

```python
# df é seu DataFrame (a tabela de dados que você carregou com o read_excel, por exemplo).
# .describe() é um método do Pandas que gera estatísticas descritivas dos dados numéricos da tabela.
# O que ele retorna:
# Uma tabela com várias informações estatísticas de cada coluna numérica do seu DataFrame.
# serve para Ver rapidamente a distribuição dos dados.Encontrar valores muito altos ou baixos (outliers).
# Ter uma ideia da tendência central (média, mediana) e da dispersão (desvio padrão).
# Saber se tem valores faltantes (olhando o count)

df.describe()
```

```python
# df é o seu DataFrame (a tabela de dados).
# .info() é um método que mostra informações gerais sobre o DataFrame, como estrutura, tipos de dados e valores nulos.
# Serve para ver se há valores ausentes (valores faltando).
# Confirmar os tipos de dados das colunas (às vezes o Pandas interpreta algo errado).
# Entender a estrutura geral da tabela antes de trabalhar com ela.
df.info()
```

```python
#Esse comando é muito usado para limpar os dados, e vamos explicar direitinho o que ele está fazendo:
# O que significa (parte por parte):
# 1. df['DATA INICIAL']
# Isso acessa a coluna chamada 'DATA INICIAL' do DataFrame.
# 2. .notna()
# Esse método retorna uma série booleana (True ou False) dizendo quais linhas têm valor e quais estão vazias (NaN) nessa coluna.
# 3. df[df['DATA INICIAL'].notna()]
# Esse comando está filtrando o DataFrame, mantendo apenas as linhas onde a coluna 'DATA INICIAL' NÃO está vazia.
# 4. df =
# Está sobrescrevendo o DataFrame original, ou seja, salvando apenas as linhas válidas (sem dados faltando nessa coluna).
# 5. print(df)
# Mostra o novo DataFrame (sem valores nulos na coluna 'DATA INICIAL').
#  Serve para Limpar os dados, eliminando linhas que não têm data na coluna 'DATA INICIAL'.
# Isso evita problemas em análises, gráficos ou cálculos que precisam dessa informação.

df = df[df['DATA INICIAL'].notna()]
print(df)
```

```python
#Esse trecho de código serve para criar um gráfico de barras personalizado usando os dados de um DataFrame (df) — provavelmente com informações sobre postos de combustíveis pesquisados em diferentes estados.
# 1. plt.figure(figsize=(12, 6))
# Cria uma nova área de desenho (figura) para o gráfico.
# figsize=(12, 6) define o tamanho da figura em polegadas → 12 de largura, 6 de altura.
# Isso ajuda a deixar o gráfico mais largo e legível, especialmente se houver muitos estados.
# 2. sns.barplot(x='ESTADO', y='NÚMERO DE POSTOS PESQUISADOS', data=df)
# Cria o gráfico de barras com a biblioteca Seaborn.
# x='ESTADO': usa os nomes dos estados no eixo X (horizontal).
# y='NÚMERO DE POSTOS PESQUISADOS': mostra a quantidade de postos no eixo Y (vertical).
# data=df: os dados vêm do DataFrame df (onde estão as colunas mencionadas).
# 3. plt.title('Quantidade de Postos Pesquisados por Estado')
# Adiciona um título ao gráfico.
# 4. plt.xlabel('Estado')
# Define o nome do eixo X como “Estado”.
# 5. plt.ylabel('Número de Postos Pesquisados')
# Define o nome do eixo Y como “Número de Postos Pesquisados”.
# 6. plt.xticks(rotation=45, ha='right')
# Gira os rótulos do eixo X (os nomes dos estados) em 45 graus para evitar sobreposição.
# ha='right' → alinha os textos à direita, o que ajuda na legibilidade.
# 7. plt.tight_layout()
# Ajusta automaticamente os espaços internos do gráfico, para garantir que nada fique cortado (como o título ou os rótulos dos eixos).
# 8. plt.show()
# Exibe o gráfico na tela
# Esse código cria um gráfico de barras com:
# Eixo X: Estados
# Eixo Y: Quantidade de postos pesquisados
# Estética ajustada para ficar mais legível e visualmente organizado

plt.figure(figsize=(12, 6))
sns.barplot(x='ESTADO', y='NÚMERO DE POSTOS PESQUISADOS', data=df)
plt.title('Quantidade de Postos Pesquisados por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Postos Pesquisados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Esse comando está fazendo duas coisas principais:
# Calculando a média do preço de revenda por estado.
# Criando um gráfico de pizza (gráfico de setores) para mostrar a distribuição do preço médio entre os estados.
#Parte 1 – Agrupando, calculando e ordenando os dados --> (preco_medio_por_estado = df.groupby('ESTADO')['PREÇO MÉDIO REVENDA'].mean().sort_values(ascending=False))
# df.groupby('ESTADO'): agrupa os dados por estado.
# ['PREÇO MÉDIO REVENDA']: seleciona a coluna com os preços médios de revenda.
# .mean(): calcula a média desses preços para cada estado.
# .sort_values(ascending=False): ordena do maior para o menor preço médio.
# Resultado: você tem uma série com o preço médio de revenda por estado, em ordem decrescente.
#Parte 2 – Criando o gráfico de pizza --> (plt.figure(figsize=(8, 8)))
#Cria uma nova figura com tamanho 8x8 polegadas.--> plt.pie(preco_medio_por_estado, labels=preco_medio_por_estado.index, autopct='%1.1f%%', startangle=140)
# Cria o gráfico de pizza com os dados calculados.
# preco_medio_por_estado: são os valores usados nos pedaços da pizza.
# labels=preco_medio_por_estado.index: define os nomes de cada fatia (os estados).
# autopct='%1.1f%%': mostra a porcentagem com uma casa decimal.
# startangle=140: gira o início do gráfico para melhor posicionamento visual.

preco_medio_por_estado = df.groupby('ESTADO')['PREÇO MÉDIO REVENDA'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 8))
plt.pie(preco_medio_por_estado, labels=preco_medio_por_estado.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição do Preço Médio de Revenda por Estado')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# Esse trecho de código está preparando uma tabela com o preço médio de revenda por município, ordenada do mais caro para o mais barato. Vamos explicar linha por linha com calma:
# Linha 1: Agrupar, calcular média e ordenar --> preco_medio_municipio = df.groupby('MUNICÍPIO')['PREÇO MÉDIO REVENDA'].mean().sort_values(ascending=False)
# O que faz:
# df.groupby('MUNICÍPIO'): agrupa os dados por município.
# ['PREÇO MÉDIO REVENDA']: seleciona a coluna de preço médio de revenda.
# .mean(): calcula a média dos preços de revenda em cada município.
# .sort_values(ascending=False): ordena os municípios do maior preço médio para o menor.
#  Resultado: uma série (lista indexada) onde:
# o índice são os nomes dos municípios,
# os valores são os preços médios de revenda.
# Linha 2: Transformar a série em DataFrame --> preco_medio_df = preco_medio_municipio.reset_index()
# O que faz:
# Transforma a série preco_medio_municipio em um DataFrame com duas colunas: 'MUNICÍPIO' e 'PREÇO MÉDIO REVENDA'.
# O índice anterior (os nomes dos municípios) vira uma coluna normal.
# Linha 3: Renomear as colunas --> preco_medio_df.columns = ['MUNICÍPIO', 'PREÇO_MÉDIO_REVENDA']
# O que faz:
# Troca os nomes das colunas para facilitar o uso (sem acentos e com underline).
# Agora o DataFrame tem:
# 'MUNICÍPIO'
# 'PREÇO_MÉDIO_REVENDA'
#Linha 4: Visualizar os primeiros dados --> preco_medio_df.head()
# O que faz:
# Mostra as 5 primeiras linhas do DataFrame.
# Isso é útil para verificar se os dados foram organizados e processados corretamente.
#  Em resumo:
# Esse conjunto de comandos:
# Calcula o preço médio de revenda por município.
# Ordena do mais caro para o mais barato.
# Organiza os dados em forma de tabela limpa.
# Mostra as primeiras linhas para conferência.

preco_medio_municipio = df.groupby('MUNICÍPIO')['PREÇO MÉDIO REVENDA'].mean().sort_values(ascending=False)
preco_medio_df = preco_medio_municipio.reset_index()
preco_medio_df.columns = ['MUNICÍPIO', 'PREÇO_MÉDIO_REVENDA']
preco_medio_df.head()
```

```python
#Esse comando é super simples, mas muito útil quando você quer destacar o município com o maior e o menor preço médio de revenda. Vamos entender linha por linha:
#  Linha 1: media_preco_alto = preco_medio_df.iloc[0]
#  O que faz:
# preco_medio_df é o DataFrame com os preços médios por município, já ordenado do maior para o menor (como vimos antes).
# .iloc[0] seleciona a primeira linha do DataFrame, ou seja, o município com o maior preço médio de revenda.
# Essa linha inteira (com nome do município e valor) é armazenada na variável media_preco_alto.
#  Linha 2: media_preco_baixo = preco_medio_df.iloc[-1]
#  .iloc[-1] pega a última linha do DataFrame, ou seja, o município com o menor preço médio de revenda.
# Essa linha também é salva, agora na variável media_preco_baixo.


media_preco_alto = preco_medio_df.iloc[0]
media_preco_baixo = preco_medio_df.iloc[-1]
```

```python
# Esse comando é bem simples, mas muito útil, especialmente quando você está explorando ou tratando um conjunto de dados no Pandas.
# print(preco_medio_df.columns)
# O que significa:
# preco_medio_df é um DataFrame (uma tabela de dados).
# .columns acessa os nomes das colunas dessa tabela.
# print(...) exibe essa informação na tela.
# Para que serve:
# Serve para ver quais colunas existem no seu DataFrame.
# Isso é útil quando:
# Você quer saber como chamar cada coluna corretamente.
# Está manipulando os dados e precisa evitar erros de digitação.
# Está explorando um arquivo novo e não sabe o que ele contém.
#Se o DataFrame tem duas colunas chamadas 'MUNICÍPIO' e 'PREÇO_MÉDIO_REVENDA', esse comando vai exibir:Index(['MUNICÍPIO', 'PREÇO_MÉDIO_REVENDA'], dtype='object')
# Essa estrutura Index(...) é como o Pandas mostra a lista de colunas.
# Em resumo:
# Esse comando mostra quais são as colunas do seu DataFrame, e te ajuda a entender como os dados estão organizados — essencial para fazer gráficos, filtros, agrupamentos etc.

print(preco_medio_df.columns)
```

```python
# Esse comando está criando um gráfico de barras horizontal com os municípios no eixo Y e os preços médios de revenda no eixo X. É uma forma clara e visual de mostrar quais municípios têm os preços mais altos ou mais baixos.
# Vamos explicar linha por linha:
# 1. plt.figure(figsize=(12, 6))
# Cria uma nova figura (área para desenhar o gráfico).
# Define o tamanho da figura: 12 polegadas de largura por 6 de altura.
# Isso ajuda o gráfico a ficar mais espaçado e legível.
# 2. sns.barplot(x='PREÇO_MÉDIO_REVENDA', y='MUNICÍPIO', data=preco_medio_df, palette='viridis')
# Usa o Seaborn para criar um gráfico de barras.
# x='PREÇO_MÉDIO_REVENDA': valores no eixo X (horizontal).
# y='MUNICÍPIO': nomes dos municípios no eixo Y (vertical).
# data=preco_medio_df: os dados estão no DataFrame preco_medio_df.
# palette='viridis': define a paleta de cores do gráfico (um gradiente bonito do roxo ao amarelo).
# Dica: como o eixo Y é o dos municípios, o gráfico vai sair na forma horizontal, o que facilita a leitura dos nomes (principalmente se houver muitos).
# 3. plt.title('Preço Médio de Revenda por MUNICÍPIO (Maiores e Menores Preços Destacados)')
# Adiciona um título ao gráfico.
# 4. plt.xlabel('Preço Médio de Revenda')
# Define o nome do eixo X.
# 5. plt.ylabel('MUNICÍPIO')
# Define o nome do eixo Y.
# Em resumo:
# Esse código gera um gráfico de barras horizontal que mostra:
# Cada município listado no eixo Y.
# O preço médio de revenda correspondente no eixo X.
# As cores ajudam a destacar variações nos preços entre os municípios.

plt.figure(figsize=(12, 6))
sns.barplot(x='PREÇO_MÉDIO_REVENDA', y='MUNICÍPIO', data=preco_medio_df, palette='viridis')
plt.title('Preço Médio de Revenda por MUNICÍPIO(Maiores e Menores Preços Destacados)')
plt.xlabel('Preço Médio de Revenda')
plt.ylabel('MUNICÍPIO')
```
